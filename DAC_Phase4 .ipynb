```python
#identifying the outliers in the dataset
numeric_columns = df1.select_dtypes(include=[np.number])

# Define a function to find outliers using IQR
def find_outliers_iqr(column):
    Q1 = np.percentile(column, 25)
    Q3 = np.percentile(column, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (column < lower_bound) | (column > upper_bound)

# Find and store outliers for each column
outliers = {}
for column in numeric_columns:
    outliers[column] = find_outliers_iqr(df1[column])

# Display or work with the outlier information
for column, is_outlier in outliers.items():
    print(f'Column: {column}')
    print(f'Number of Outliers: {is_outlier.sum()}')
    print(df1[is_outlier])
```
